```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/aidencamilleri/Downloads/2023DetailedList.csv")
```

```python
df.head()
```

```python
for i in range(0, len(df.Totaldue)):
    df.Totaldue[i] = df.Totaldue[i].lstrip("$")
for j in range(0, len(df.Appraisal)):
    df.Appraisal[j] = df.Appraisal[j].lstrip("$")
df.head()
```

```python
for k in range(0,len(df.City)):
    if df.City[k] != "CHARLESTON":
        df.drop(index=[k], inplace=True)
df.head()
```

```python
df
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="RPTax")
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
df["Address"] = df["SitusAddr"] + ", Charleston County, SC"
df.head()
```

```python
df["location"] = df["Address"].apply(geocode)
```

```python
dfdropped = df.location.dropna()
dfdropped.head()
```

```python
len(dfdropped)
```

```python
location = geolocator.geocode(df.SitusAddr[0] + ", " + df.City[0])
print(location.address)
print(location.latitude)
print(location.longitude)
```
